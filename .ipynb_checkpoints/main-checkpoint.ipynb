```python
# Number of Divisions 
yDiv = 10 
zDiv = 10 
```

```python
# Establishing Deltas 
yDel = (df[1].max()-df[1].min())/yDiv   # yDel
zDel = (df[2].max()-df[2].min())/zDiv   # zDel
```

```python
# Establish Lower and Upper Bounds
yLb = df[1].min()
yUb = df[1].min()+ yDel

zLb = df[2].min()
zUb = df[2].min()+ zDel
```

```python
# y splits 
dfAr = list()     # List of y splits 
for i in range(yDiv):
    df1 = df[(df[1] >= yLb) & (df[1] <= yUb)]
    dfAr.append( df1 )
    yLb = yUb
    yUb = yUb + yDel
```

```python
# Check for Y split wastage 
suma = 0 
for split_index in range(len(dfAr)):
    suma = suma + len(dfAr[split_index])
```

```python
dfAr2 = list()
for i in dfAr:
    print('len:',len(i))
    zLb = df[2].min()
    zUb = df[2].min()+ zDel

    for j in range(zDiv):
        
        df2 = i[(i[2] <= zUb)&(i[2] >= zLb) ]
        zLb = zUb 
        zUb = zUb+ zDel
        dfAr2.append(df2)
        
    
    
    
```
