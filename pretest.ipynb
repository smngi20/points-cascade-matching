```python
# FILTERING FOR LOW REFLECTION VALUES 
df3 = src2[src2['146.000000'] > 50]  
```

```python
new = pd.DataFrame(columns=['0','1','2','3','4','5','6','7'])
```

```python
new
```

```python
import math 
from tqdm import tqdm 

flag = 0 
for index, row in tqdm(df3.iterrows()): 
    flag = flag + 1 
#     if flag>22220:
#         break
 
    x = row["4.89099979"]
    y = row["-0.50099999"]
    z = row["0.60399997"]

    # Magnitude
    mag =  math.sqrt(x**2+y**2+z**2)
    
    if flag%2==0 :
        actual_mag = mag + STD
    else :
        actual_mag = mag - STD

    # Unit Vector 
    u_x = x/mag
    u_y = y/mag
    u_z = z/mag
    
    u_x * actual_mag

    test = {"0":u_x * actual_mag,
             "1":   u_y * actual_mag,
             "2":   u_z * actual_mag,
              "3":  row["238"],
              "4":  row["255"],
              "5":  row["0"],
              "6":  row["1.000000"],
              "7":  row["146.000000"],
                
           }
    


    
    new = new.append(test,ignore_index = True)
    
    
```

```python
new.to_csv('output.csv', index=False)
```
